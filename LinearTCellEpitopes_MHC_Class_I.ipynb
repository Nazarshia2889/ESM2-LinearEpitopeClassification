```python
# Import data
df = pd.read_csv('tcellmhc1.csv')
```

```python
columnNames = df.columns

for val in range(len(df.iloc[0, :])):
  print(columnNames[val], ": ", df.iloc[0, val])
```

```python
df_filter = df[['Epitope - IEDB IRI', 'Epitope - Object Type', 'Epitope - Name', 'Epitope - Source Molecule', 'Epitope - Molecule Parent IRI',
               'Epitope - Molecule Parent', 'Epitope - Source Organism', 'Effector Cell - Name', 'MHC Restriction - Class', 
               'Antigen Presenting Cell - Name', 'Assay - Qualitative Measurement']]

df_filter
```

```python
df_filter.isna().sum()
```

```python
df_filter = df_filter[df_filter['Epitope - Molecule Parent IRI'].notna()]
```

```python
df_filter['Effector Cell - Name'].unique()
```

```python
df_filter['MHC Restriction - Class'].unique()
```

```python
df_filter['Epitope - Object Type'].unique()
```

```python
df_filter.isnull().sum()
```

```python
df_filter = df_filter[(df_filter['Effector Cell - Name'].notna()) & ~df_filter['Effector Cell - Name'].str.contains('Unknown/Unspecified', na=False) & ~df_filter['Effector Cell - Name'].str.contains('Other', na=False) & ~df_filter['Effector Cell - Name'].str.contains('B cell', na=False)]
```

```python
df_filter['ID'] = df_filter['Epitope - IEDB IRI'].apply(lambda x : x[(len(x) - x[::-1].index('/')):])
```

```python
df_filter.rename(columns = {'Epitope - Name': 'Sequence'}, inplace = True)
# df_filter.rename(columns = {'Epitope - IEDB IRI': 'ID'}, inplace = True)
```

```python
df_filter
```

```python
df_filter[df_filter['ID'] == '742288']
```

```python
# positive/negative to numerical
df_filter['target'] = df_filter['Assay - Qualitative Measurement'].apply(lambda x: {'Negative': -1, 'Positive-Low': 0.9, 'Positive-High': 1.1, 'Positive-Intermediate': 1, 'Positive': 1}[x])
```

```python
df_filter['Parent ID'] = df_filter['Epitope - Molecule Parent IRI'].apply(lambda x : x[(len(x) - x[::-1].index('/')):])
```

```python
df_filter.head()
```

```python
df_filter['Effector Cell - Name'].value_counts()
```

```python
# combine into unique epitopes
df_new = df_filter.drop_duplicates(subset='ID').merge(df_filter.groupby('ID')['target'].aggregate(lambda x: 1 if x.sum() > 0 else 0 if x.sum() != 0 else np.nan), left_on='ID', right_index=True, how='left').rename(columns={'target_y': 'target'}).drop('target_x', axis=1).dropna(subset=['target'])
df_new
```
