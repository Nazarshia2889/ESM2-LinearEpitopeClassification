```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# original_df = pd.read_csv('bcell.csv')
original_df = pd.read_feather('bcell.feather')
df = original_df.copy()
```

```python
original_df
```

```python
df = original_df[['Epitope - IEDB IRI', 'Epitope - Object Type', 'Epitope - Name', 'Epitope - Source Molecule', 'Epitope - Molecule Parent IRI', 'Epitope - Molecule Parent', 'Epitope - Source Organism', 'Assay - Qualitative Measure']]
df
```

```python
df['Epitope - IEDB IRI'] = df['Epitope - IEDB IRI'].apply(lambda x: x.split('/')[-1])
```

```python
# rename columns
df.rename(columns={'Epitope - Name': 'Sequence'}, inplace=True)
df.rename(columns={'Epitope - IEDB IRI': 'ID'}, inplace=True)
```

```python
# positive/negative to numerical
df['target'] = df['Assay - Qualitative Measure'].apply(lambda x: {'Negative': -1, 'Positive-Low': 0.9, 'Positive-High': 1.1, 'Positive-Intermediate': 1, 'Positive': 1}[x])
```

```python
# drop when Epitope - Molecule Parent IRI is None
df = df[df['Epitope - Molecule Parent IRI'].notna()]

# add parent ID
df['Parent ID'] = df['Epitope - Molecule Parent IRI'].apply(lambda x: x[(len(x) - x[::-1].index('/')):])
```

```python
# create combined ID
df['Combined ID'] = df['ID'] + '_' + df['Parent ID']
```

```python
# combine into unique epitopes
df = df.drop_duplicates(subset='Combined ID').merge(df.groupby('Combined ID')['target'].aggregate(lambda x: 1 if x.sum() > 0 else -1 if x.sum() != 0 else np.nan), left_on='Combined ID', right_index=True, how='left').rename(columns={'target_y': 'target'}).drop('target_x', axis=1).dropna(subset=['target'])
```

```python
# switch target from -1/1 to 0/1
df['target'] = df['target'].apply(lambda x: 0 if x < 0 else 1)
```

```python
df.to_csv('cleaned/bcell.csv', index=False)
df.reset_index().drop('index', axis=1).to_feather('cleaned/bcell.feather')
```
